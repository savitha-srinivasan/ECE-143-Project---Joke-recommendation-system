```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
import seaborn as sns
```

```python
df=pd.read_csv('TheAmountOfJokesRatedByEachUser.csv', sep=',',header=None)
```

```python
x=np.array(df[0][1:].values,dtype=int)
y=np.array(df[1][1:].values,dtype=int)
```

### The amount of jokes rated by user

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 12))
ax.plot(x, y)
ax.set_title("The amount of jokes rated by user")
ax.set_xlabel("User's ID")
ax.set_ylabel("Users rate how many jokes")
```

### The number of words in each joke

```python
l=list()
with open('/home/w8lin/Data/Joke.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        l.append(len(row[1]))
l=l[1:]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 12))
ax.plot(np.arange(0,len(l)), l)
ax.set_title("The number of words in each joke")
ax.set_xlabel("Joke's ID")
ax.set_ylabel("Numbers of words in this joke")
```

### The Average Rating on each Jokes

```python
df = pd.read_csv('JokeForEachRating.csv')
df= df.loc[:, df.columns != "Unnamed: 0"]
```

```python
jokeAvgsRating = np.nanmean(df, axis=0)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
ax.bar(np.arange(0,len(jokeAvgsRating)), jokeAvgsRating)
ax.set_title("Average Rating of Jokes")
ax.set_xlabel("Joke's ID")
ax.set_ylabel("Average Rating")
```

### The user's average rating on every joke

```python
userAvgsRating = np.nanmean(df, axis=1)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(20, 10))
ax.bar(np.arange(0,len(userAvgsRating)), userAvgsRating)
plt.ylim(-25, 20)
ax.set_title("User's average rating on every joke")
ax.set_xlabel("User's ID")
ax.set_ylabel("Average Rating for every joke")
```

### Distribution of joke scores

```python
ave_jokes = df.iloc[:,1:].aggregate(axis=0, func='mean').sort_values()
joke_values = ave_jokes.dropna()
sns.distplot(joke_values)
plt.xlabel('Average score')
plt.ylabel('Ratio of jakes')
plt.title('Distribution of joke scores')
```
