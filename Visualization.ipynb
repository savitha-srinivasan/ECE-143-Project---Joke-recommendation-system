```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
```

```python
df=pd.read_csv('TheAmountOfJokesRatedByEachUser.csv', sep=',',header=None)
```

```python
x=np.array(df[0][1:].values,dtype=int)
y=np.array(df[1][1:].values,dtype=int)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 12))
ax.plot(x, y)
ax.set_title("The amount of jokes rated by each user")
ax.set_xlabel("User's ID")
ax.set_ylabel("Users rate how many jokes")
```

```python
l=list()
with open('/home/w8lin/Data/Joke.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        l.append(len(row[1]))
l=l[1:]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 12))
ax.plot(np.arange(0,len(l)), l)
ax.set_title("The number of words in each joke")
ax.set_xlabel("Joke's ID")
ax.set_ylabel("Numbers of words in this joke")
```
